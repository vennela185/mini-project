Project title

Student Marks Analysis Using Python and NumPy

Problem Statement

The objective of this program is to create a Python application that collects marks of multiple students, stores them using a NumPy array, and performs basic calculations such as total, average, highest, and lowest marks. Based on the average marks, the program assigns a grade using conditional statements. The program uses basic Python concepts like functions, loops, lists, and NumPy operations.


Logic Explanation

The program begins by defining a user-defined function in which all operations are performed. An empty list is first created to store the marks of students. Using a for loop, the program takes marks as input from the user and stores them in the list. This list is then converted into a NumPy array to perform mathematical operations easily. The total and average marks are calculated using NumPy functions. Based on the calculated average, conditional statements are used to assign a grade to the students. Finally, the calculated results and grade are displayed using print statements.

```python
import numpy as np

def marks_analysis():
    marks_list = []

    for i in range(3):
        marks = int(input("Enter marks: "))
        marks_list.append(marks)

    marks_array = np.array(marks_list)

    total = marks_array.sum()
    average = marks_array.mean()

    if average > 85:
        grade = "A"
    elif average >= 70:
        grade = "B"
    elif average >= 50:
        grade = "C"
    else:
        grade = "Fail"

    print("Total:", total)
    print("Average:", average)
    print("Grade:", grade)

marks_analysis()
```
